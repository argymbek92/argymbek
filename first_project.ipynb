Задание 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? Не стоит забывать, что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть.

"""
Задание 5. Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц. 

1.Когорты строил на время создания заказа order_purchase_timestamp
2.Создал новый столбец year_month, который является годом и месяцем, когда произошел первый заказ клиента
3.Агрегировал клиентов, покупки и суммы покупок по каждой year_month с помощью min()
4.Создал сводную таблицу с помощью pivot_table
5.Выявил когорту с самым высоким retention на 3й месяц
"""

```python
res_task_5 = df[['customer_unique_id', 'order_purchase_timestamp']]
```

```python
res_task_5
```

```python
# Поскольку я делаю ежемесячные когорты, я буду смотреть на общее ежемесячное поведение наших клиентов.
# Создадим новый столбец year_month, который является годом и месяцем, когда произошел первый заказ клиента.
res_task_5['year_month'] = res_task_5.order_purchase_timestamp.dt.strftime('%Y-%m')
```

```python
res_task_5
```

```python
# Теперь, нам необходимо агрегировать клиентов, покупки и суммы покупок по каждой year_month
cohorts = res_task_5.groupby(['customer_unique_id','order_purchase_timestamp'], as_index=False).agg({'year_month': 'min'}).rename(columns={'year_month': 'cohortgroup'})
```

```python
cohorts['ret_month'] = cohorts.groupby('customer_unique_id', as_index=False).cohortgroup.cumcount()
```

```python
cohorts['retention'] = 1
```

```python
cohorts
```

```python
# Создаю сводную таблицу с помощью pivot_table
cohorts = pd.pivot_table(cohorts, index='cohortgroup', columns='ret_month', values='retention', aggfunc=np.sum, fill_value=0)
```

```python
cohorts
```

```python
# В период с января по декабрь  с самым высоким retention на 3й месяц. 
cohorts[3].max()
```

"""
Задание 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.


1.Выбрать методику расчета RFM-кластеров
2.Я выбрал простейшую для 1 варианта:
    R - разница между текущей датой и датой последнего заказа (в днях с последнего заказа)
    F - количество заказов за все время
    M - общая сумма покупок за все время
3.Кластеризация: разбиение по квартилям этих метрик
4.Рассчитал время от каждого заказа до максимально возможного времени внутри датасета
5.Рассчитал стоимость заказа для каждого заказа в данных
6.Для каждого пользователя вычислил:
    срок в днях от его последнего заказа
    суммарное количество заказов
    общую сумму заказов
7.Выставил оценки, разбивая на диапазон от 1 до 5

"""

```python
# Я буду опираться на order_purchase_timestamp —  время создания заказа, как и в предыдущих заданиях. 
df.head()
```

```python
# Вместо текущей даты будем использовать max+1. 
from datetime import timedelta
```

```python
last_date = df['order_purchase_timestamp'].max() + timedelta(days=1)
```

```python
# Для создания recency, frequency и monetary я группирую df по customer_id.
rfm = df.reset_index().groupby('customer_id')\
        .agg({'order_purchase_timestamp': lambda x: (last_date - x.max()).days,
             'order_id': 'count',
             'price': 'sum'})
```

```python
rfm.rename(columns={'order_purchase_timestamp': 'recency',
                   'order_id': 'frequency',
                   'price': 'monetary'}, inplace=True)
```

```python
rfm
```

```python
# Теперь нам нужно выставить оценки. 
# Мы будем разбивать на диапазон от 1 до 4. 
#Чем шире диапазон, тем более точные наши группы, но в тоже время работать с большим количествов комбинаций сложнее.
quantiles = rfm.quantile(q=[0.2, 0.4, 0.6, 0.8])
```

```python
quantiles
```

```python
quantiles = quantiles.to_dict()
```

```python
quantiles
```

```python
def get_segment(x,p,d):
    if x <= d[p][0.2]:
        return 1
    elif x <= d[p][0.4]:
        return 2
    elif x <= d[p][0.6]:
        return 3
    elif x <= d[p][0.8]:
        return 4
    else:
        return 5
```

```python
rfm['R'] = rfm['recency'].apply(get_segment, args=('recency', quantiles,))
```

```python
rfm['F'] = rfm['frequency'].apply(get_segment, args=('frequency', quantiles,))
```

```python
rfm['M'] = rfm['monetary'].apply(get_segment, args=('monetary', quantiles,))
```

```python
rfm['RFM_Score'] = rfm.R.map(str) + rfm.F.map(str) + rfm.M.map(str)
```
